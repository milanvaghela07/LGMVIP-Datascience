## **Modifying the original data**

```python
# renaming the columns

df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
df.head()
```

```python
new_data=df[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

## **Data Exploration**

```python
# plotting heat map

plt.figure(figsize=(9,6))
new_data.corr()
sns.heatmap(new_data.corr(), annot=True)
```

```python
# plotting the histogram

new_data.hist(figsize=(20,20))
plt.show()
```

## **Terrorist activities in each region through area plot**

```python
# a crosstab is a table showing the relationship between two or more variables.
# plotting area plot graph

pd.crosstab(new_data.Year, new_data.Region).plot(kind='area',figsize=(12,5))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.grid()
plt.show()
```

## **Counting terrorist activities each year through countplot** 

```python
# plotting count graph

plt.subplots(figsize=(15,5))
sns.countplot(x='Year',data = new_data)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.grid()
plt.show()
```

## **Calculating the percentage of attacks that has happened between 1970 and 2017**

```python
Year=new_data.Year.value_counts().to_dict()
rate=((Year[1979]-Year[1970])/Year[1979])*100
print(Year[1970],'attacks happened in 1970 !')
print(Year[1979],'attacks happened in 1979 !')
print('Total number of attacks from 1970 has increased by',np.round(rate,0),'% till 1979 !')
```

## **Different types of attacks**

```python
plt.figure(figsize=(16,3))
sns.countplot(x = new_data['AttackType'], order = new_data['AttackType'].value_counts().index)
plt.xticks(rotation=70)
plt.xlabel('Method')
plt.title('Method of Attack')
plt.show()
```

## **Different types of targets**

```python
plt.figure(figsize=(16,5))
sns.countplot(x = new_data['Target_type'], order=new_data['Target_type'].value_counts().index)
plt.xticks(rotation=70)
plt.xlabel('Type')
plt.title('Type of Target')
plt.grid()
plt.show()
```

## **Destructive Feature of data**

```python
#Destructive Feature of data

print("Country with the most attacks:",new_data['Country'].value_counts().idxmax())
print("City with the most attacks:",new_data['city'].value_counts().index[1]) #as first entry is 'unknown'
print("Region with the most attacks:",new_data['Region'].value_counts().idxmax())
print("Year with the most attacks:",new_data['Year'].value_counts().idxmax())
print("Month with the most attacks:",new_data['Month'].value_counts().idxmax())
print("Group with the most attacks:",new_data['Group'].value_counts().index[1])
print("Most Attack Types:",new_data['AttackType'].value_counts().idxmax())
```

## **Top Countries affected by Terror Attacks**

```python
# Total Number of people killed in terror attack
killData = new_data.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))# drop the NaN values
```

```python
# Let's look at what types of attacks these deaths were made of.
attackData = new_data.loc[:,'AttackType']
# attackData
typeKillData = pd.concat([attackData, killData], axis=1)
typeKillData.head()
```

## **Count of terrorist attacks in each countries and regions**

```python
fig,axes = plt.subplots(figsize=(16,5),nrows=1,ncols=2)
sns.barplot(x = new_data['Country'].value_counts()[:20].values, y = new_data['Country'].value_counts()[:20].index, ax=axes[0]);
axes[0].set_title('Terrorist Attacks per Country')
sns.barplot(x=new_data['Region'].value_counts().values, y=new_data['Region'].value_counts().index, ax=axes[1])
axes[1].set_title('Terrorist Attacks per Region')
fig.tight_layout()
plt.show()
```
